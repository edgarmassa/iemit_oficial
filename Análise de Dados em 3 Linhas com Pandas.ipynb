```python
import pandas as pd
```

```python
vendas = pd.DataFrame({
    'Produto': ['Notebook', 'Monitor', 'Mouse', 'Teclado', 'Notebook', 'Monitor', 'Mouse'],
    'Categoria': ['Eletrônicos', 'Eletrônicos', 'Acessórios', 'Acessórios', 'Eletrônicos', 'Eletrônicos', 'Acessórios'],
    'Valor': [4500, 1200, 80, 120, 4600, 1150, 75]
})
vendas
```

```python
analise = vendas.groupby(['Categoria', 'Produto']).agg({'Valor': ['sum', 'mean', 'min', 'max']})
analise
```
