```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [1500, 1800, 2200, 1900, 2800, 3100]

plt.plot(meses, vendas, marker='o', color='#0066cc', linewidth=2)

plt.title('Vendas por Mês - 1° Semestre')
plt.xlabel('Meses')
plt.ylabel('Vendas (R$)')

plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
